### 🛠️ Verilog AI Assistant
This assistant will take your design specifications and generate synthesizable Verilog code using the Gemini API.

````python
# 1. Setup: Install and import libraries
!pip install -q -U google-generativeai

import google.generativeai as genai
from google.colab import userdata
try:
    GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
    genai.configure(api_key=GOOGLE_API_KEY)
except:
    print('Please add your GOOGLE_API_KEY to the Colab Secrets (🔑) to proceed.')
````

````python
# 2. Define the Assistant Function
def generate_verilog_assistant(design_name, inputs, outputs, specs, extras=""):
    model = genai.GenerativeModel('gemini-1.5-flash')

    prompt = f"""
    You are an expert Hardware Description Language (HDL) engineer.
    Generate a professional, synthesizable Verilog module based on these specs:

    - Module Name: {design_name}
    - Inputs: {inputs}
    - Outputs: {outputs}
    - Functional Specifications: {specs}
    - Additional Requirements: {extras}

    Provide ONLY the Verilog code inside a markdown code block.
    """

    response = model.generate_content(prompt)
    full_text = response.text

    # Extract code between ```verilog and ``` if present
    import re
    code_match = re.search(r'```(?:verilog)?(.*?)```', full_text, re.DOTALL)
    verilog_code = code_match.group(1).strip() if code_match else full_text.strip()

    # Save to file
    filename = f"{design_name}.v"
    with open(filename, "w") as f:
        f.write(verilog_code)

    print(f"Code generated and saved to {filename}\n")
    print("--- Generated Verilog ---")
    print(verilog_code)
````

### Fill in your design details below:

````python
# @title Design Input Form
DESIGN_NAME = "alu_8bit" # @param {type:"string"}
INPUTS = "clk, reset, [7:0] a, [7:0] b, [2:0] opcode" # @param {type:"string"}
OUTPUTS = "reg [7:0] result, reg carry_out" # @param {type:"string"}
SPECS = "Perform ADD, SUB, AND, OR, XOR based on opcode. Include synchronous reset." # @param {type:"string"}
EXTRAS = "Use a case statement and optimize for area." # @param {type:"string"}

generate_verilog_assistant(DESIGN_NAME, INPUTS, OUTPUTS, SPECS, EXTRAS)
````
